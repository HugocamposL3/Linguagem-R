```r
install.packages("deldir")
library(deldir)
```

```r
# Data
set.seed(1)
x <- runif(50)
y <- runif(50)

# Calculate Voronoi Tesselation and tiles
tesselation <- deldir(x, y)
tiles <- tile.list(tesselation)

plot(tiles, pch = 19)
```

Se você quiser, você pode adicionar números em vez dos pontos padrão, configurando .

```r
# Data
set.seed(1)
x <- runif(50)
y <- runif(50)

# Calculate Voronoi Tesselation and tiles
tesselation <- deldir(x, y)
tiles <- tile.list(tesselation)

plot(tiles, pch = 19,
     number = TRUE)
```

```r
# Data
set.seed(1)
x <- runif(50)
y <- runif(50)

# Calculate Voronoi Tesselation and tiles
tesselation <- deldir(x, y)
tiles <- tile.list(tesselation)

plot(tiles,
     showpoints = FALSE)
```

As bordas das telhas podem ser fechadas com um ajuste retângulo

```r
# Data
set.seed(1)
x <- runif(50)
y <- runif(50)

# Calculate Voronoi Tesselation and tiles
tesselation <- deldir(x, y)
tiles <- tile.list(tesselation)

plot(tiles,
     close = TRUE)
```

ocê pode passar uma paleta de cores para o argumento da função de plotagem para adicionar cores aos azulejos. Observe que você também pode alterar a cor ou os pontos com ou a cor dos números com . Este último no caso 

```r
# Data
set.seed(1)
x <- runif(50)
y <- runif(50)

# Calculate Voronoi Tesselation and tiles
tesselation <- deldir(x, y)
tiles <- tile.list(tesselation)

plot(tiles, pch = 19,
     fillcol = hcl.colors(50, "Purple-Yellow"))
```

O argumento permite mudar a cor da borda das telhas,como mostra o exemplo abaixo.

```r
# Data
set.seed(1)
x <- runif(50)
y <- runif(50)

# Calculate Voronoi Tesselation and tiles
tesselation <- deldir(x, y)
tiles <- tile.list(tesselation)

plot(tiles, pch = 19,
     border = "white",
     fillcol = hcl.colors(50, "Sunset"))
```

A função também fornece um método de recorte com . O argumento precisa de uma lista com os componentes e dar as coordenadas do polígono para o qual a tessellation deve ser cortada. No exemplo a seguir, estamos cortando-o em um círculo.

```r
install.packages('polyclip')
```

```r
# Data
set.seed(1)
x <- runif(50)
y <- runif(50)

# Calculate Voronoi Tesselation and tiles
tesselation <- deldir(x, y)
tiles <- tile.list(tesselation)

# Circle
s <- seq(0, 2 * pi, length.out = 3000)
circle <- list(x = 0.5 * (1 + cos(s)),
               y = 0.5 * (1 + sin(s)))
#plot (titles, pch = 19,col.pts = "white",border = "white", fillcol = hcl.colors(50,"viridis"),clipp = circle)
plot(tiles, pch = 19, col.pts = "white", border = "white", fillcol = hcl.colors(50, "viridis"),clipp = circle)
```
